```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('HR_comma_sep.csv')
data.head()
```

```python
tr_data=data[['satisfaction_level','average_montly_hours','promotion_last_5years','salary','Department']]
```

```python

tr_data
```

```python
tr_data.drop('Department',axis='columns',inplace=True)
```

```python
tr_data
```

```python
dummy=pd.get_dummies(tr_data[['salary']],prefix='salary')
dummy
```

```python
dummy['salary_high']=dummy['salary_high'].replace(True,1)
dummy['salary_high']=dummy['salary_high'].replace(False,0)
dummy['salary_low']=dummy['salary_low'].replace(True,1)
dummy['salary_low']=dummy['salary_low'].replace(False,0)
dummy['salary_medium']=dummy['salary_medium'].replace(False,0)
dummy['salary_medium']=dummy['salary_medium'].replace(True,1)
dummy
```

```python
tr_data2=pd.concat([tr_data,dummy],axis='columns')
tr_data2
```

```python
tr_data2.drop('salary',axis='columns',inplace=True)
tr_data2
```
