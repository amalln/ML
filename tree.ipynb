```python
import pandas as pd
```

```python
data=pd.read_csv('titanic.csv')
data.head()
```

```python
data1=data.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
data1
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data2=data1[['Pclass','Sex','Age','Fare']]
data2
```

```python

le_s=LabelEncoder()


```

```python

data2['sexx']=le_s.fit_transform(data2['Sex'])
data2
```

```python
data2=data2.drop(['Sex'],axis='columns')
```

```python
data2
```

```python
data3=data['Survived']
data3
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data2,data3,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
cls=DecisionTreeClassifier()
cls.fit(x_train,y_train)
```

```python
cls.predict(x_test)
```

```python
cls.score(x_test,y_test)
```
